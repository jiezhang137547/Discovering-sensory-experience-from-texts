```python
import pandas as pd
import re
```

```python
df=pd.read_csv("raw_tweets.csv")
```

```python
# pattern1 = r"#\w+"
p4 = re.compile(r'(?:@\s*)((\w+ ){3})(?=https)?', re.IGNORECASE)
y = [p4.search(t) for t in df['tweet']]

```

```python
def match(matches):
    matcheds = []
    for i in range(len(matches)):
        if matches[i] is not None:
            matched = matches[i].group(0)
            matcheds.append(matched)
        else:
            matcheds.append(None)
    return matcheds
```

```python
match = match(y)
```
